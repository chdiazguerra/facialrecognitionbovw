```python
from src.features.feature_extraction import *
import yaml
```

```python
IMG_PATH = "../data/pain/f10a1.jpg"

METHOD = "sift"
CONFIG_FILE = "../src/features/config.yaml"
```

```python
with open(CONFIG_FILE, "r", encoding="utf-8") as file:
    config = yaml.safe_load(file)
```

```python
config_method = config.get(METHOD, None)

if config_method is None:
    raise Exception("No existe la configuracion para el metodo escogido")

if METHOD=="sift":
    extractor = SIFT(**config_method)
if METHOD=="brisk":
    extractor = BRISK(**config_method)
if METHOD=="orb":
    extractor = ORB(**config_method)
if METHOD=="daisy":
    extractor = DAISY(**config_method)
```
